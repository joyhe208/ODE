<h3>Direction Field along with initial conditions</h3>
$\,t=0\,,y_{0}=5$

```python
#assuming r=.1
def diff(y):
    return .1*y
```

```python
t = np.linspace(0,15,15)
y = np.linspace(0,30,30)
y_ex = 5*np.e**(.1*t)

plt.figure(figsize=(8,8))
for time in t:
    for popsize in y:
        slope = diff(popsize)
        xvals = np.linspace(time-.12,time+.12,2)
        z = slope*(xvals-time)+popsize
        plt.plot(xvals,z)
plt.plot(t[:],y_ex[:])
```
