<h3>Direction Field along with initial conditions</h3>

```python
#assuming r=.1
def diff(y):
    return .1*y
```

```python
t = np.linspace(0,15,15)
y = np.linspace(0,30,30)
y_ex = 5*np.e**(.1*t)
y_ex2 = 3*np.e**(.1*t)
y_ex3 = np.zeros(len(t))
y_ex4 = 7*np.e**(.1*t)
plt.figure(figsize=(8,8))
for time in t:
    for popsize in y:
        slope = diff(popsize)
        xvals = np.linspace(time-.12,time+.12,2)
        z = slope*(xvals-time)+popsize
        plt.plot(xvals,z)
plt.plot(t,y_ex)
plt.plot(t,y_ex2)
plt.plot(t,y_ex3)
plt.plot(t,y_ex4)
```
