```python
import numpy as np
import matplotlib.pyplot as plt
```

<h1>Logistic Growth with a Threshold</h1>
<br>
$$
\frac{dy}{dt}=ry(\frac{y}{T}-1)(1-\frac{y}{N})
$$

```python
threshold = 10
capacity = 20
growth = .1
t = np.linspace(0,30,30)
y = np.linspace(0,30,30)
def diff(T,N,r,pop):
    return r*pop*((pop/T)-1)*(1-(pop/N))
```

```python
def plotNum(y_0):
    ys = []
    for i in range(len(t)):
        if(i==0):
            ys.append(y_0)
        else:
            ys.append(diff(threshold, capacity, growth,ys[i-1])+ys[i-1])
    plt.plot(t, ys)
```

```python
y_01 = 25
y_02 = 15
y_03 = 7
plt.figure(figsize=(15,15))
for time in t:
    for popsize in y:
        slope = diff(threshold, capacity, growth, popsize)
        xvals = np.linspace(time-.12,time+.12,2)
        z = slope*(xvals-time)+popsize
        plt.plot(xvals,z)
plotNum(y_01)
plotNum(y_02)
plotNum(y_03)
```
